```python
import pandas as pd
import numpy as np
```

```python
def Precision(eval,emo,epsilon ):
    
    total = 0
    print('===============Precision=============')
    for i in emo :
        retrieved = len(eval[eval['dominant']== i])
        relevant = len(eval[(eval['dominant']==i)&(eval['real']==i)])
        total += relevant/(retrieved+epsilon)
        print(i,'retrieved: ', retrieved, 'relevant:', relevant,"precision:", round(relevant/(retrieved+epsilon),2))
        
    
    print("\nAverage Precision: ", round(total/8,2))
    
```

```python
def Recall(eval,emo,epsilon ):
    total = 0
    print('\n===============Recall=============')
    for i in emo :
        relevant = len(eval[eval['real']== i])
        retrieved = len(eval[(eval['dominant']==i)&(eval['real']==i)])
        total += retrieved/(relevant+epsilon)
        print(i,'relevant: ', relevant, 'retrieved:', retrieved,"Recall:", round(retrieved/(relevant+epsilon),2))
        
    
    print("\nAverage Recall: ", round(total/8,2))
```

```python
def MAP(eval,emo):
    print('===============MAP=============')
    print(emo)
    retrieved = eval[eval['dominant']== emo]
    retrieved_rank = retrieved.sort_values(by = [emo],axis=0,ascending = False)
    count = 0
    total = 0
    min = len(retrieved_rank) if 10>len(retrieved_rank) else 10
    iter = 1
    for i,row in retrieved_rank.iterrows():
        if row['real']==emo:
            count+=1
            total += count/iter  
            mark = 'O'
        else:
            mark = 'X'
  
        print(iter," - ",mark, 'precision :', round(count/iter,2),'(', count,'/',iter,')')
        iter+=1
    
    
    print("\nMAP : ",round(total/count,2))
        
                     
```

```python
eval = pd.read_excel("data/evaluation_2016_2.xlsx")
emotions = ['Anger','Anticipation','Disgust','Fear','Joy','Sadness','Surprise','Trust','Love']
epsilon = 0.00000001
Precision(eval,emotions,epsilon)
Recall(eval, emotions,epsilon )
MAP(eval,'Anticipation')
MAP(eval,'Joy')
MAP(eval,'Sadness')
MAP(eval,'Love')

```
