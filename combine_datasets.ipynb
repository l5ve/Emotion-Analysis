```python
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
for year in range(2006, 2020):
    data = pd.read_csv("./Analysis_Result_2/result_"+str(year)+" (1).csv")
    data=data[['title','artist','year','dominant']]
    df=df.append(data,ignore_index=True)

df.rename({'dominant': 'sentiment'}, axis=1)
```

```python
df.to_csv("./User Emotion/data/song_lyrics.csv",index=False)
```
